```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('train_plots/training_03-01-21_08-17') as f:
    content = f.readlines()

content = [x.strip().split(' ') for x in content] 
content = [float(x) for sublist in content for x in sublist]
content = np.array(content).reshape((-1, 4))
```

```python
train_loss_hist = content[:, 0]
test_loss_hist = content[:, 1]
train_acc_hist = content[:, 2]
test_acc_hist = content[:, 3]
```

```python
fig, ax1 = plt.subplots(figsize=(5, 4))


ax1.set_xlabel('Epochs', size=13)
ax1.set_ylabel('Cross entropy loss', size=13)
ax1.plot(train_loss_hist, color='lightskyblue', label='train loss')
ax1.plot(test_loss_hist, color='steelblue', label='test loss')
ax1.set_ylim([0,4])
ax1.legend(loc='upper left')


ax2 = ax1.twinx()

ax2.set_ylabel('Accuracy', size=13)
ax2.plot(train_acc_hist, color='sandybrown', label='train acc')
ax2.plot(test_acc_hist, color='sienna', label='test acc')
ax2.set_ylim([0,1])


ax2.legend(loc='upper right')
plt.savefig('visual_training.png')

ax2.set_xlim([0, 100])


plt.show()
```
